    Comparing with old and new features adding to the existing product.
    To know the better quality from existing old feature or newly added feature results,
    We apply Hypothesis testing.

    It is a part of Stastical Analysis, where we test the assumptions made regarding a population parameter.
    It is generally used when we were to compare a single group with an external standard and two or more groups with each other.


    Hypothesis testing is done on sample data.

## Null Hypothesis Testing
    That suggests there is no statistical significance exists betweeen the populations. It is denoted by H0 and read as H-naught

    product A = product B
    the testing from two products results same.
    H0 must always contain equality (=)
    
## Alternative Hypothesis
    That suggests there is a statistical significance exists betweeen the populations parameters.
    It is the contrast of the Null hypothesis. It is denoted by H1 and read as H-one

    product A != product B (or)
    product A < product B (or)
    product A > product B

    H1 must always contain (!=, <, >)
    
    the testing from two products results same.


## Types of Hypothesis Testing 
    Z-Test:              
        - for Average on a different situations
        - both mean and standard deviation from population data 
        - sample data (which must be >30)
        
    T-Test:        
        for Average on a another situations
    Chi-square Test: 
        good and realtion between two products


## Steps of Hypothesis Testing 
    - State null(H0) and alternative(H1) hypothesis
    - Choose level of significance(alpha )
    - find critical values
    - find test statistic
    - draw your conclusions
