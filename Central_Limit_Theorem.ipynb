### the graph tells, not a normal distribution from the population data

### now, converting population data into normal distribution
    
    the length of the population data is 1000.
    pick 10% of sample data from population data.
    i,e (1000*10)/100 = 100

```python
(1000*10)/100
```

```python
np.random.choice(pop_data)
```

```python
sample_mean = []
for num_sample in range(50):   # increase the sample data and check the normal distribution
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
len(sample_mean) # collected 50 mean values from 50 samples data
```

```python
sample_mean[:5]  # to show only 5 mean values from 50 mean values in a list
```

```python
sample_m = pd.DataFrame({"Sample_Mean":sample_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample_Mean", data=sample_m)
plt.show()
```

### the graph tells, a normal distribution from a sample data (50)

```python
sample_mean = []
for no_sample in range(100):   # increase more sample data, getting the normal distribution better
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
sample_m = pd.DataFrame({"Sample_Mean":sample_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample_Mean", data=sample_m)
plt.show()
```

### the graph tells, a normal distribution from a sample data (100)

```python
np.mean(pop_data), np.mean(sample_mean)  # central limit theorem
```

    Both population and sample data will be having almost equal mean values
    When mean cannot be taken from population data then pick sample data and find mean.

# This is central limit theorem

```python
def testDec(func):
    
    def wrapper():
        print("before the function is called.")
        func()
        print("after the function is called.")
    return wrapper
    
@testDec 
def sayHello():
    print("output")

sayHello()
```

```python
def my_decorator(func):
    # function within function
    def wrapper():
        print("before the function is called.")
        func()
        print("after the function is called.")
    return wrapper   # return the function wrapper

@my_decorator
def say_hello():
    print("Output!")

say_hello()
```
