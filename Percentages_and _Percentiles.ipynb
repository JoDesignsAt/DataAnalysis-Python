# Percentages and Percentiles

        Percentages: 
            ex: the individual person subject marks - this have base value
                sci - 80, math - 80, soc - 80, bio - 80, eng - 80, from 100
                80*5 = (400/500)*100 = 80% percentage from total marks
                
        Percentiles: 
            ex: A rank from a group  - no base value
                100 students in a class, you got 4th rank
                (100 - 4/100 )*100 = 96% percentile ranking from total students

                percentiles starts from - 0% to 100%

```python
(100 - 4)/100*100
```

# Interquartile Range (IQR)
           using IQR, we can find outliers

           IQR = Q3 - Q1

           by Box ploting

    By distributing percentiles into 4 equal parts - 
    
        1st quartile as 25% percentile - minimum
        2nd quartile as 50% percentile - median
        3rd quartile as 75% percentile - maximum

        percentile starts from 0 - 100 at the end
        

# Outliers 
    The values which are out of range are called outliers
    
            list = [2,3,4,7,8,12,18]  the values seems in the range
            list = [2,3,4,7,8,12,18,500,1000] the values 500,1000 are out of range
     

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("titanic.csv")
```

```python
dataset.head(5)
```

# Percentiles

```python
np.percentile(dataset["age"], 25)  # 25 is a 25% percentile
```

    null values in dataset["age"]

```python
dataset["age"].value_counts()
```

    count null values

```python
dataset["age"].isnull().sum()
```

    remove 177 null values using fillna()

```python
dataset["age"].fillna(dataset["age"].mean(), inplace=True)
```

    check any null?

```python
dataset["age"].isnull().sum()
```

    None, apply percentile

```python
np.percentile(dataset["age"], 25) 
```

     IQR range
            Q1 - 25 percentile, Q3 - 75 percentile

```python
np.percentile(dataset["age"], 25) , np.percentile(dataset["age"], 75)  # Q1 - 25 percentile, Q3 - 75 percentile
```

        the MIN and MAX of the data

```python
np.percentile(dataset["age"], 0) , np.percentile(dataset["age"], 100) 
```

    check all with .describe()

```python
dataset["age"].describe()
```

# Boxplotting 

```python
sns.boxplot(x="age", data=dataset)
plt.show()
```

### IQR boxplotting helps to find outliers and remove outliers from the dataset
