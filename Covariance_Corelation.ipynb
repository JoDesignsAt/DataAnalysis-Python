```python
dataset.select_dtypes(np.number).cov()
```

    A high covariance basically indicates there is a strong relationship between the variables. A low value means there is a weak relationship.

```python
data_corr = dataset.select_dtypes(np.number).corr()
data_cov = dataset.select_dtypes(np.number).cov()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_corr, annot=True) # annot to enable values
plt.show()
```

    the lighter shade, the better the relation.
    1 indicates the stronger relation between two columns.
    color black indicates 0.
    Usually in Machine learning, correlation is used to check the relation between two columns.

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_cov, annot=True) # annot to enable values
plt.show()
```

    The higher value indicates strong relationhip between two columns.
    Covariance is usually used for feature selection or for simple regression.
    when there is to check relation between two columns, always correlation is used.
    itâ€™s a measure that indicates the extent to which two random variables change in cycle. 

    The covariance between two random variables is a measure of how the two variables move together.
    A positive covariance indicates that the random variables tend to move together. 
    A negative covariance indicates that when one variable is above its mean the other tends to be below its mean.
    
    Correlation coefficient is some sort of a standard form of covariance that measures the strength of linear relationship among the two variables, which isn't captured by covariance alone. 
    
    Covariance is a measure and co-efficient of correlation is a ratio.

    Covariance is for two random variables whereas Variance is for one random variable. Co-variance tells how two series are related.

# Co-variance and Correlation

a. If co-variance = 0, then the two series are not correlated (independent of each other)

b. If co-variance > 0, then the two series are positively correlated.

c. If co-variance < 0, then the two series are negatively correlated (inverse dependence).


    So basically Co-variance divided by the Standard deviations of the two series = Correlation.
