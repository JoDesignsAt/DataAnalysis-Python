# Mean Absolute Deviation
    MAD of a dataset is the average distance between each data point and the mean
    It tells variability in a dataset, how much it is spreading, low spread is good for analysis

```python
# to check which session of students will perform in greater

ses_a = np.array([75,65,73,68,72,76]) # marks of students of session-A
ses_b = np.array([90,47,43,96,93,51]) # marks of students of session-B

y = np.array([1,2,3,4,5,6])
```

```python
a_mean = np.array([75,65,73,68,72,76]).mean()
```

```python
b_mean = np.array([90,47,43,96,93,51]).mean()
```

```python
a_mean, b_mean 
# But the mean of two dataset are equal or almost same.
# Here, we have to choose one dataset bewtween two, then check the spreadness of the data.
# how to choose? check the data points which are close to mean line, i,e low spread, the performance will be good
# else large spread, data points are away from mean line, which not good for analysis
# use Mean Absolute Deviation method, to find the spreadness
```

```python
plt.figure(figsize=(10,5))
plt.scatter(ses_a,y,label="ses_a",color="blue")
plt.scatter(ses_b,y,label="ses_b",color="red")
plt.legend() # to show labels
plt.show()
```

```python
mean = np.array([75,65,73,68,72,76]).mean()
```

```python
mad_a = np.sum(np.abs(ses_a - mean))/len(ses_a) 
```

```python
mad_b = np.sum(np.abs(ses_b - mean))/len(ses_b)
```

```python
mad_a, mad_b
```

```python
plt.figure(figsize=(10,5)) # size of the figure 

plt.scatter(ses_a,y,label="ses_a",color="blue")
plt.scatter(ses_b,y,label="ses_b",color="red")

plt.plot([70,70,70,70,70,70],y, c="green", label="mean") # mean line
#plt.plot([mean for i range(1,8],y,c="green", label="mean")

plt.legend() # to show labels
plt.show()
```

# Standard deviation * Standard deviation  = variance
    the relationship between STANDARD DEVIATION and VARIANCE

```python
ses_a = np.array([75,65,73,68,72,76]) # marks of students of session-A
ses_b = np.array([90,47,43,96,93,51]) # marks of students of session-B
```

#### Standard deviation

```python
np.std(ses_a) , np.std(ses_b)
```

#### Variance

```python
np.var(ses_a) , np.var(ses_b)
```

```python
np.std(ses_a) * np.std(ses_a)  # this is equal to variance
```

    Like low 'MEAN' value is considered, also with low 'VARIANCE AND STANDARD DEVIATION' value is considered
    therefore, consider np.var(ses_a) which has low value
    Becuase the spreadness is large in high value, and with low value the spreadness is close
