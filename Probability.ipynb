# Random variable
    A random variable x is a function, that assigns a real number to each outcome in the sample space of a random experiment.

# Probability
    A probaility measures the likelihood of a particular outcome or event occuring.
        P(A) = no. of times A occurs / total no. of possible outcomes.

    Ex: 70% chances to get rain, there is a probility of getting rain.
        how much/many chances of getting is something probability.
        
# Probability distribution
    It describes how the probabilities of different outcomes are distributed over the sample space of a random variable.

    - Discrete probabilty distributions
    - Continuous probabilty distributions
